```python
# PREPARING THE DATA FOR MODEL

# Transform data in the standard scaler machine learning model, so the columns should have values which can be used in machine learning models. 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
StandardScaler = StandardScaler()
columns_to_scale = ['age','trestbps', 'chol', 'thalach', 'oldpeak']
heart_data[columns_to_scale] = StandardScaler.fit_transform(heart_data[columns_to_scale])
```

```python
# To observe the transformed values after applying the standard scalar
heart_data.head()
```

```python
# X will have all the attributes, but the target is dropped since we have train the data
# y will have only the target.

X= heart_data.drop(['target'], axis=1)
y= heart_data['target']
```

```python
X_train, X_test,y_train, y_test=train_test_split(X,y,test_size=0.3,random_state=40)
```

```python
# CHECK THE SAMPLE SIZE
# training data is 70% and testing data is 30%

print('X_train-', X_train.size)
print('X_test-',X_test.size)
print('y_train-', y_train.size)
print('y_test-', y_test.size)
```

```python
# LOGISTIC REGRESSION

from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()

model1=lr.fit(X_train,y_train)
prediction1=model1.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix

cm=confusion_matrix(y_test,prediction1)
cm
```

```python
sns.heatmap(cm, annot=True,cmap='BuPu')
```

```python
TP=cm[0][0]
TN=cm[1][1]
FN=cm[1][0]
FP=cm[0][1]
print('Testing Accuracy:',(TP+TN)/(TP+TN+FN+FP))
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,prediction1)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, prediction1))
```

```python
# DECISION TREE

from sklearn.tree import DecisionTreeClassifier

dtc=DecisionTreeClassifier()
model2=dtc.fit(X_train,y_train)
prediction2=model2.predict(X_test)
cm2= confusion_matrix(y_test,prediction2)

cm2
```

```python
accuracy_score(y_test,prediction2)
```

```python
print(classification_report(y_test, prediction2))
```

```python
# RANDOM FOREST

from sklearn.ensemble import RandomForestClassifier

rfc=RandomForestClassifier()
model3 = rfc.fit(X_train, y_train)
prediction3 = model3.predict(X_test)
confusion_matrix(y_test, prediction3)
```

```python
accuracy_score(y_test, prediction3)
```

```python
print(classification_report(y_test, prediction3))
```

```python
from sklearn.svm import SVC

svm=SVC()
model4=svm.fit(X_train,y_train)
prediction4=model4.predict(X_test)
cm4= confusion_matrix(y_test,prediction4)

cm4
```

```python
accuracy_score(y_test, prediction4)
```

```python
from sklearn.naive_bayes import GaussianNB

NB = GaussianNB()
model5 = NB.fit(X_train, y_train)
prediction5 = model5.predict(X_test)
cm5= confusion_matrix(y_test, prediction5)

cm5
```

```python
accuracy_score(y_test,prediction5)
```
